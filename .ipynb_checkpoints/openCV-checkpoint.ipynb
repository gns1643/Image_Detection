```python
import cv2
import numpy as np
from rembg import remove
from PIL import Image
import io
import svgwrite
from ipywidgets import interact, IntSlider
import matplotlib.pyplot as plt
import cv2
```

```python
# [Î∏îÎ°ù 2] Ïù¥ÎØ∏ÏßÄ Î°úÎìú Î∞è Î∞∞Í≤Ω Ï†úÍ±∞ Ìï®Ïàò
def load_and_remove_bg(path):
    print(f"üîÑ [1Îã®Í≥Ñ] Î∞∞Í≤Ω Ï†úÍ±∞ ÏãúÏûë: {path}")
    
    # 1. ÌååÏùº ÏùΩÍ∏∞ Î∞è rembgÎ°ú Î∞∞Í≤Ω Ï†úÍ±∞
    with open(path, 'rb') as i:
        input_data = i.read()
        output_data = remove(input_data)
    
    # 2. PIL Ïù¥ÎØ∏ÏßÄÎ°ú Î≥ÄÌôò
    img_pil = Image.open(io.BytesIO(output_data))
    
    # 3. Ìà¨Î™ÖÌïú Î∂ÄÎ∂Ñ(Alpha)ÏùÑ Ìù∞ÏÉâÏúºÎ°ú Î≥ÄÍ≤Ω (Ï§ëÏöî!)
    # Ìà¨Î™ÖÎèÑÎ•º Ìù∞ÏÉâÏúºÎ°ú Ï±ÑÏö∞ÏßÄ ÏïäÏúºÎ©¥ ÎÇòÏ§ëÏóê Í≤ÄÏùÄÏÉâÏúºÎ°ú Ïù∏ÏãùÎêòÏñ¥ ÌÖåÎëêÎ¶¨Í∞Ä ÏÉùÍ∏∏ Ïàò ÏûàÏùå
    background = Image.new("RGB", img_pil.size, (255, 255, 255))
    try:
        background.paste(img_pil, mask=img_pil.split()[3]) # Alpha Ï±ÑÎÑê ÏÇ¨Ïö©
    except IndexError:
        # ÏïåÌåå Ï±ÑÎÑêÏù¥ ÏóÜÎäî Ïù¥ÎØ∏ÏßÄÏùº Í≤ΩÏö∞ Í∑∏ÎÉ• Î∂ôÏûÑ
        background.paste(img_pil)
        
    # 4. OpenCV Ìè¨Îß∑(Numpy array)ÏúºÎ°ú Î≥ÄÌôò
    img_cv = np.array(background)
    img_cv = cv2.cvtColor(img_cv, cv2.COLOR_RGB2BGR)
    
    print("‚úÖ Î∞∞Í≤Ω Ï†úÍ±∞ ÏôÑÎ£å")
    
    # Í≤∞Í≥º ÌôïÏù∏ (JupyterÏö©)
    plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
    
    return img_cv

# Ïã§Ìñâ 
img_processed = load_and_remove_bg(input("ÌååÏùº Î™ÖÏùÑ ÏûÖÎ†•ÌïòÏÑ∏Ïöî (ÌôïÏû•Ïûê Ìè¨Ìï® ÌïÑÏöî) : "))
```

```python
#ÏÇ¨ÏßÑ Ï∂úÎ†• Ìï®Ïàò(ÌÖåÏä§ÌåÖÏö©)
def printImage(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

```python
# [Î∏îÎ°ù 3] Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Î∞è Î∏îÎü¨ÎßÅ
def preprocess_image_gray(img_cv):
    print("üîÑ [2-1Îã®Í≥Ñ] Ï†ÑÏ≤òÎ¶¨(Gray) ÏãúÏûë")
    
    # Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Î≥ÄÌôò
    gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
    
    
    
    print("‚úÖ Ï†ÑÏ≤òÎ¶¨ ÏôÑÎ£å")

    printImage(gray)
    return gray
def preprocess_image_blurr(img_cv):
    print("üîÑ [2-2Îã®Í≥Ñ] Ï†ÑÏ≤òÎ¶¨(Blur) ÏãúÏûë")
    # Í∞ÄÏö∞ÏãúÏïà Î∏îÎü¨ (ÎÖ∏Ïù¥Ï¶à Ï†úÍ±∞)
    # (5, 5)Îäî Ïª§ÎÑê ÌÅ¨Í∏∞ÏûÖÎãàÎã§. Ïà´ÏûêÍ∞Ä ÌÅ¥ÏàòÎ°ù Îçî Î≠âÍ∞úÏßëÎãàÎã§.
    blurred = cv2.GaussianBlur(img_cv, (5, 5), 0)
    print("‚úÖ Ï†ÑÏ≤òÎ¶¨ ÏôÑÎ£å")

    printImage(blurred)
    return blurred
# Ïã§Ìñâ 
img_gray = preprocess_image_gray(img_processed)
img_blurr = preprocess_image_blurr(img_gray)
```

```python
import matplotlib.pyplot as plt
import cv2

# [Ï§ëÏöî] img_blurredÍ∞Ä ÏóÜÏúºÎ©¥ ÏóêÎü¨Í∞Ä ÎÇòÎãà, ÌòπÏãú ÏóêÎü¨ÎÇòÎ©¥ Ïïû Îã®Í≥Ñ(Î∏îÎ°ù 3)Î•º Î®ºÏ†Ä Ïã§ÌñâÌï¥Ï£ºÏÑ∏Ïöî.
# img_processed = load_and_remove_bg("input.png")
# img_blurred = preprocess_image(img_processed)

def show_grid_search():
    print("üîÑ Ïó¨Îü¨Í∞ÄÏßÄ ÏÑ§Ï†ïÍ∞íÏùÑ ÌÖåÏä§Ìä∏ÌïòÎäî Ï§ë...")
    
    # ÌÖåÏä§Ìä∏Ìï† ÏÑ§Ï†ïÍ∞í ÏÑ∏Ìä∏ (ÎßàÏùåÏóê ÎìúÎäî Ï°∞Ìï©Ïù¥ ÏóÜÏúºÎ©¥ Ïù¥ Ïà´ÏûêÎ•º Î∞îÍøîÎ≥¥ÏÑ∏Ïöî)
    # (th1, th2) Ï°∞Ìï© Î¶¨Ïä§Ìä∏
    settings = [
        (30, 100), (50, 150), (80, 180),
        (30, 80),  (50, 100), (80, 150),
        (20, 50),  (40, 80),  (60, 120)
    ]
    
    plt.figure(figsize=(12, 12)) # Ï†ÑÏ≤¥ Í∑∏Î¶º ÌÅ¨Í∏∞
    
    for i, (th1, th2) in enumerate(settings):
        # 1. Ïó£ÏßÄ Í≤ÄÏ∂ú
        edges = cv2.Canny(img_blurr, th1, th2)
        
        # 2. 9Í∞ú Ï§ë ÌïòÎÇòÎ°ú Î∞∞Ïπò (3x3 Í≤©Ïûê)
        plt.subplot(3, 3, i + 1)
        plt.imshow(edges, cmap='gray')
        plt.title(f'No.{i+1} : ({th1}, {th2})') # Î≤àÌò∏ÏôÄ ÏÑ§Ï†ïÍ∞í ÌëúÏãú
        plt.axis('off')
        
    plt.tight_layout()
    plt.show()

# Ïã§Ìñâ
show_grid_search()
```

```python
import cv2
import svgwrite
import numpy as np

# ÏïÑÍπå Ï†ïÌïú ÏµúÏ†ÅÏùò Í∞í
FINAL_TH1 = 50
FINAL_TH2 = 100

def generate_all_files(img_blurr, nc_filename="drawing.nc", svg_filename="output.svg"):
    print(f"üîÑ ÌååÏùº ÏÉùÏÑ± ÏãúÏûë (Threshold: {FINAL_TH1}, {FINAL_TH2})")
    
    # 1. Ïó£ÏßÄ Í≤ÄÏ∂ú
    edges = cv2.Canny(img_blurr, FINAL_TH1, FINAL_TH2)
    contours, _ = cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    
    # 2. SVG Ï§ÄÎπÑ (Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ ÌôïÏù∏)
    height, width = img_blurr.shape
    dwg = svgwrite.Drawing(svg_filename, profile='tiny', size=(width, height))
    
    # 3. G-ÏΩîÎìú ÏÑ§Ï†ïÍ∞í (Î°úÎ¥áÏóê ÎßûÏ∂∞ ÏàòÏ†ï Í∞ÄÎä•)
    Z_SAFE = 10.0   # Ïù¥Îèô ÎÜíÏù¥
    Z_DRAW = 0.0    # Í∑∏Î¶¨Í∏∞ ÎÜíÏù¥
    FEED_RATE = 1000 
    SCALE = 0.2     # 1ÌîΩÏÖÄ = 0.2mm (ÌÅ¨Í∏∞ Ï°∞Ï†àÏö©)
    
    # 4. G-ÏΩîÎìú ÌååÏùº Ïó¥Í∏∞
    with open(nc_filename, 'w') as f:
        # G-ÏΩîÎìú Ìó§Îçî ÏûëÏÑ±
        f.write("%\n")
        f.write("G21 (Units: mm)\nG90 (Absolute)\n")
        f.write(f"G0 Z{Z_SAFE}\n") # ÏïàÏ†Ñ ÎÜíÏù¥Î°ú Îì§Í∏∞
        
        count = 0
        
        for contour in contours:
            # Ïû°Ìã∞ Ï†úÍ±∞ (Í∏∏Ïù¥ 15 Ïù¥ÌïòÎäî Î¨¥Ïãú)
            if cv2.arcLength(contour, closed=False) < 15:
                continue
            
            # Îã®ÏàúÌôî (Ï†ê Í∞úÏàò Ï§ÑÏù¥Í∏∞)
            epsilon = 0.002 * cv2.arcLength(contour, closed=False)
            approx = cv2.approxPolyDP(contour, epsilon, closed=False)
            
            # --- [SVG Ï†ÄÏû•] ---
            # SVGÎäî Ï¢åÌëú Î≥ÄÌôò ÏóÜÏù¥ Í∑∏ÎåÄÎ°ú Ï†ÄÏû• (ÏõπÎ∏åÎùºÏö∞Ï†Ä ÌôïÏù∏Ïö©)
            points = [(float(p[0][0]), float(p[0][1])) for p in approx]
            dwg.add(dwg.polyline(points, stroke='black', fill='none', stroke_width=2))
            
            # --- [G-ÏΩîÎìú Ï†ÄÏû•] ---
            # ÏãúÏûëÏ†êÏúºÎ°ú Ïù¥Îèô
            start_p = approx[0][0]
            sx, sy = start_p[0] * SCALE, start_p[1] * SCALE
            
            # Ï£ºÏùò: Î°úÎ¥á Ï¢åÌëúÍ≥ÑÏóê ÎßûÏ∂∞ YÏ∂ïÏóê ÎßàÏù¥ÎÑàÏä§(-)Î•º Î∂ôÏûÑ
            f.write(f"G0 X{sx:.3f} Y{-sy:.3f}\n") 
            f.write(f"G1 Z{Z_DRAW} F{FEED_RATE}\n") # Ìéú ÎÇ¥Î¶¨Í∏∞
            
            for i in range(1, len(approx)):
                p = approx[i][0]
                px, py = p[0] * SCALE, p[1] * SCALE
                f.write(f"G1 X{px:.3f} Y{-py:.3f}\n")
            
            f.write(f"G0 Z{Z_SAFE}\n") # Ìöç ÎÅùÎÇòÎ©¥ Ìéú Îì§Í∏∞
            count += 1
            
        # G-ÏΩîÎìú ÎßàÎ¨¥Î¶¨
        f.write("G0 X0 Y0\nM2\n%\n")
        
    # SVG ÌååÏùº Ï†ÄÏû• ÏôÑÎ£å
    dwg.save()

    print(f"‚úÖ ÏÉùÏÑ± ÏôÑÎ£å! Ï¥ù {count}Í∞úÏùò Ìöç")
    print(f"   1Ô∏è‚É£ SVG ÌååÏùº: {svg_filename} (ÌÅ¨Î°¨ÏúºÎ°ú Ïó¥Ïñ¥ÏÑú ÌôïÏù∏)")
    print(f"   2Ô∏è‚É£ NC  ÌååÏùº: {nc_filename}  (Î°úÎ¥áÏóêÍ≤å Ï†ÑÎã¨Ìï† ÌååÏùº)")

# Ïã§Ìñâ
generate_all_files(img_blurr)
```

```python
#Ï§ëÏã¨ÏÑ† Ï∂îÏ∂ú Î∞©Ïãù
import cv2
import svgwrite
import numpy as np
from skimage.morphology import skeletonize # ÌïµÏã¨ ÎùºÏù¥Î∏åÎü¨Î¶¨ Ï∂îÍ∞Ä

def generate_skeleton_files(img_blurr, nc_filename="drawing_skeleton.nc", svg_filename="output_skeleton.svg"):
    print(f"üîÑ Ï§ëÏã¨ÏÑ† Ï∂îÏ∂ú Î∞©Ïãù(Skeletonization)ÏúºÎ°ú Î≥ÄÌôò ÏãúÏûë...")

    # [1Îã®Í≥Ñ] Ïù¥ÏßÑÌôî (Binarization)
    # Ïù¥ÎØ∏ÏßÄÎ•º ÏôÑÏ†ÑÌûà Í≤ÄÏùÄÏÉâ(0)Í≥º Ìù∞ÏÉâ(255)ÏúºÎ°ú ÎÇòÎàïÎãàÎã§.
    # Canny ÎåÄÏã† Adaptive ThresholdÎ•º ÏÇ¨Ïö©ÌïòÏó¨ 'Î©¥'ÏùÑ Ïû°ÏäµÎãàÎã§.
    # 11: Ï£ºÎ≥Ä ÌîΩÏÖÄ Î∏îÎ°ù ÌÅ¨Í∏∞ / 2: Í∞êÏÇ∞ ÏÉÅÏàò (Ï°∞Ï†à Í∞ÄÎä•)
    binary = cv2.adaptiveThreshold(img_blurr, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                   cv2.THRESH_BINARY_INV, 11, 2)

    # [2Îã®Í≥Ñ] Ïä§ÏºàÎ†àÌÜ§Ìôî (Skeletonization) - ÎºàÎåÄ ÎßåÎì§Í∏∞
    # Ìù∞ÏÉâ ÏòÅÏó≠(True)ÏùÑ 1ÌîΩÏÖÄ ÎëêÍªòÏùò ÏÑ†ÏúºÎ°ú ÍπéÏïÑÎÉÖÎãàÎã§.
    print("   ‚è≥ ÎºàÎåÄ ÍπéÎäî Ï§ë... (Ïû†Ïãú Í±∏Î¶¥ Ïàò ÏûàÏäµÎãàÎã§)")
    skeleton = skeletonize(binary > 0) # True/FalseÎ°ú Î≥ÄÌôòÌï¥ÏÑú ÎÑ£Ïñ¥Ïïº Ìï®
    
    # Îã§Ïãú OpenCVÏö© Ïù¥ÎØ∏ÏßÄ(0~255)Î°ú Î≥ÄÌôò
    skeleton_img = (skeleton * 255).astype(np.uint8)

    # [3Îã®Í≥Ñ] Ïô∏Í≥ΩÏÑ†(Ï¢åÌëú) Ï∂îÏ∂ú
    # Ïù¥Ï†ú ÏÑ†Ïù¥ 1ÌîΩÏÖÄÏù¥ÎØÄÎ°ú, ÌÖåÎëêÎ¶¨Î•º Îî∞ÎèÑ Ï§ëÏã¨ÏÑ† ÌïòÎÇòÎßå ÎÇòÏòµÎãàÎã§.
    contours, _ = cv2.findContours(skeleton_img, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    
    # ---------------------------------------------------------
    # [Ïù¥Ìïò G-ÏΩîÎìú Î∞è SVG ÏÉùÏÑ± Î°úÏßÅÏùÄ ÎèôÏùºÌï©ÎãàÎã§]
    # ---------------------------------------------------------
    
    # SVG Ï§ÄÎπÑ
    height, width = img_blurr.shape
    dwg = svgwrite.Drawing(svg_filename, profile='tiny', size=(width, height))
    
    # Î°úÎ¥á ÏÑ§Ï†ïÍ∞í
    Z_SAFE = 10.0   
    Z_DRAW = 0.0    
    FEED_RATE = 1000 
    SCALE = 0.2     
    
    # G-ÏΩîÎìú ÌååÏùº Ïì∞Í∏∞
    with open(nc_filename, 'w') as f:
        f.write("%\n")
        f.write("G21 (Units: mm)\nG90 (Absolute)\n")
        f.write(f"G0 Z{Z_SAFE}\n") 
        
        count = 0
        
        for contour in contours:
            # ÎÑàÎ¨¥ ÏßßÏùÄ Ïû°Ìã∞ Ï†úÍ±∞ (Ï§ëÏã¨ÏÑ† Î∞©ÏãùÏùÄ ÎÖ∏Ïù¥Ï¶àÍ∞Ä Ï¢Ä Îçî ÏÉùÍ∏∏ Ïàò ÏûàÏñ¥ 20ÏúºÎ°ú ÎÜíÏûÑ)
            if cv2.arcLength(contour, closed=False) < 20:
                continue
            
            # Îã®ÏàúÌôî (Ï†ê Í∞úÏàò Ï§ÑÏù¥Í∏∞)
            epsilon = 0.002 * cv2.arcLength(contour, closed=False)
            approx = cv2.approxPolyDP(contour, epsilon, closed=False)
            
            # --- SVG Í∑∏Î¶¨Í∏∞ ---
            points = [(float(p[0][0]), float(p[0][1])) for p in approx]
            dwg.add(dwg.polyline(points, stroke='black', fill='none', stroke_width=1)) # ÏñáÍ≤å ÌëúÏãú
            
            # --- G-ÏΩîÎìú Ïì∞Í∏∞ ---
            start_p = approx[0][0]
            sx, sy = start_p[0] * SCALE, start_p[1] * SCALE
            
            f.write(f"G0 X{sx:.3f} Y{-sy:.3f}\n") 
            f.write(f"G1 Z{Z_DRAW} F{FEED_RATE}\n") 
            
            for i in range(1, len(approx)):
                p = approx[i][0]
                px, py = p[0] * SCALE, p[1] * SCALE
                f.write(f"G1 X{px:.3f} Y{-py:.3f}\n")
            
            f.write(f"G0 Z{Z_SAFE}\n") 
            count += 1
            
        f.write("G0 X0 Y0\nM2\n%\n")
        
    dwg.save()

    print(f"‚úÖ ÏÉùÏÑ± ÏôÑÎ£å! Ï¥ù {count}Í∞úÏùò Ìöç")
    print(f"   1Ô∏è‚É£ SVG ÌååÏùº: {svg_filename}")
    print(f"   2Ô∏è‚É£ NC  ÌååÏùº: {nc_filename}")

# Ïã§Ìñâ (img_blurr Î≥ÄÏàòÍ∞Ä ÏûàÏñ¥Ïïº Ìï©ÎãàÎã§)
if __name__ == "__main__":
    # ÌòπÏãú img_blurrÍ∞Ä ÏóÜÎã§Î©¥ ÏïûÏÑ† Î∏îÎ°ù Ïã§Ìñâ ÌïÑÏöî
    # img_blurrÍ∞Ä Ï†ïÏùòÎêòÏñ¥ ÏûàÎã§Î©¥ Î∞îÎ°ú ÏïÑÎûò Ï§Ñ Ïã§Ìñâ
    generate_skeleton_files(img_blurr)
```
