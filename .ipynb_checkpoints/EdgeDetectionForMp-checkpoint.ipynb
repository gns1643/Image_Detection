```python
import cv2
import svgwrite
import numpy as np
import mediapipe as mp
from rembg import remove
from PIL import Image
import io
import matplotlib.pyplot as plt

# =========================================================
# [ì„¤ì •] ì¶œë ¥ ë° ë“œë¡œì‰ ê´€ë ¨ ìƒìˆ˜
# =========================================================
NC_FILENAME = "face_drawing.nc"
SVG_FILENAME = "face_drawing.svg"

# G-code ìƒì„± íŒŒë¼ë¯¸í„°
Z_SAFE = 5.0      # ì´ë™ ì‹œ ë†’ì´ (íœ ë“¤ê¸°)
Z_DRAW = 0.0      # ê·¸ë¦¬ê¸° ë†’ì´ (íœ ë‚´ë¦¬ê¸°)
FEED_RATE = 1000  # ì´ë™ ì†ë„
SCALE = 0.3       # ì´ë¯¸ì§€ ì¢Œí‘œ -> G-code ì¢Œí‘œ ìŠ¤ì¼€ì¼ (0.3ë°° ì¶•ì†Œ)

# =========================================================
# [Block 1] ì´ë¯¸ì§€ ë¡œë“œ ë° ë°°ê²½ ì œê±° (rembg)
# =========================================================
def load_and_remove_bg(path):
    """
    ì´ë¯¸ì§€ íŒŒì¼ì„ ì½ì–´ ë°°ê²½ì„ ì œê±°í•˜ê³ , íˆ¬ëª… ì˜ì—­ì„ í°ìƒ‰ìœ¼ë¡œ ì±„ìš´ ë’¤ OpenCV í¬ë§·ìœ¼ë¡œ ë°˜í™˜
    """
    print(f"ğŸ”„ [1ë‹¨ê³„] ë°°ê²½ ì œê±° ì‹œì‘: {path}")
    
    try:
        # 1. íŒŒì¼ ì½ê¸° ë° rembgë¡œ ë°°ê²½ ì œê±°
        with open(path, 'rb') as i:
            input_data = i.read()
            output_data = remove(input_data)
        
        # 2. PIL ì´ë¯¸ì§€ë¡œ ë³€í™˜
        img_pil = Image.open(io.BytesIO(output_data))
        
        # 3. íˆ¬ëª…í•œ ë¶€ë¶„(Alpha)ì„ í°ìƒ‰ìœ¼ë¡œ ë³€ê²½
        # (íˆ¬ëª…ë„ê°€ ìˆìœ¼ë©´ ë‚˜ì¤‘ì— ê²€ì€ìƒ‰ìœ¼ë¡œ ì¸ì‹ë  ìˆ˜ ìˆìœ¼ë¯€ë¡œ í°ìƒ‰ í•©ì„±ì´ ì¤‘ìš”)
        background = Image.new("RGB", img_pil.size, (255, 255, 255))
        try:
            background.paste(img_pil, mask=img_pil.split()[3]) # Alpha ì±„ë„ ì‚¬ìš©
        except IndexError:
            background.paste(img_pil) # ì•ŒíŒŒ ì±„ë„ ì—†ìœ¼ë©´ ê·¸ëƒ¥ ë¶™ì„
            
        # 4. OpenCV í¬ë§·(Numpy array)ìœ¼ë¡œ ë³€í™˜ (RGB -> BGR)
        img_cv = np.array(background)
        img_cv = cv2.cvtColor(img_cv, cv2.COLOR_RGB2BGR)
        
        print("âœ… ë°°ê²½ ì œê±° ì™„ë£Œ")
        
        # (ì˜µì…˜) ì£¼í”¼í„° ë…¸íŠ¸ë¶ ë“±ì—ì„œ í™•ì¸ìš© - í•„ìš” ì—†ìœ¼ë©´ ì£¼ì„ ì²˜ë¦¬
        # plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
        # plt.axis('off')
        # plt.show()
        
        return img_cv

    except Exception as e:
        print(f"âŒ ì´ë¯¸ì§€ ë¡œë“œ ì‹¤íŒ¨: {e}")
        return None

# =========================================================
# [Block 2] AI ì–¼êµ´ íŠ¹ì§• ì¶”ì¶œ (MediaPipe)
# =========================================================
def extract_face_contours_mediapipe(img):
    """
    OpenCV ì´ë¯¸ì§€ë¥¼ ë°›ì•„ MediaPipeë¡œ ì–¼êµ´ ì£¼ìš” ë¶€ìœ„ì˜ ì„ (Contours)ì„ ì¶”ì¶œ
    """
    print("ğŸ”„ [2ë‹¨ê³„] AI ì–¼êµ´ ëœë“œë§ˆí¬ ë¶„ì„ ì¤‘...")
    
    mp_face_mesh = mp.solutions.face_mesh
    height, width = img.shape[:2]
    
    # ì–¼êµ´ ì¸ì‹ ì„¤ì •
    with mp_face_mesh.FaceMesh(
        static_image_mode=True,     # ì •ì§€ ì´ë¯¸ì§€ ëª¨ë“œ
        max_num_faces=1,            # ì–¼êµ´ 1ê°œë§Œ ì¸ì‹
        refine_landmarks=True,      # ëˆˆë™ì ë“± ì •ë°€ ì¸ì‹
        min_detection_confidence=0.5
    ) as face_mesh:
        
        # MediaPipeëŠ” RGBë¥¼ ì‚¬ìš©í•˜ë¯€ë¡œ ë³€í™˜
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        results = face_mesh.process(img_rgb)
        
        if not results.multi_face_landmarks:
            print("âš ï¸ ì–¼êµ´ì„ ì°¾ì§€ ëª»í–ˆìŠµë‹ˆë‹¤! (ì‚¬ëŒ ì •ë©´ ì‚¬ì§„ì„ ê¶Œì¥í•©ë‹ˆë‹¤)")
            return []

        landmarks = results.multi_face_landmarks[0].landmark
        
        # ì¶”ì¶œí•  ì–¼êµ´ ë¶€ìœ„ ì •ì˜ (MediaPipe Topology)
        # ë”•ì…”ë„ˆë¦¬ì— ê° ë¶€ìœ„ì˜ ì—°ê²°ì„  ì •ë³´ë¥¼ ë‹´ìŠµë‹ˆë‹¤.
        features = {
            "lips": mp.solutions.face_mesh.FACEMESH_LIPS,
            "left_eye": mp.solutions.face_mesh.FACEMESH_LEFT_EYE,
            "right_eye": mp.solutions.face_mesh.FACEMESH_RIGHT_EYE,
            "left_eyebrow": mp.solutions.face_mesh.FACEMESH_LEFT_EYEBROW,
            "right_eyebrow": mp.solutions.face_mesh.FACEMESH_RIGHT_EYEBROW,
            "face_oval": mp.solutions.face_mesh.FACEMESH_FACE_OVAL,
            "nose": mp.solutions.face_mesh.FACEMESH_NOSE
        }
        
        final_contours = []
        
        # ê° ë¶€ìœ„ë³„ ì—°ê²°ì„ ì„ ìˆœíšŒí•˜ë©° ì¢Œí‘œ ì¶”ì¶œ
        for name, connections in features.items():
            for connection in connections:
                start_idx = connection[0]
                end_idx = connection[1]
                
                pt1 = landmarks[start_idx]
                pt2 = landmarks[end_idx]
                
                # ì •ê·œí™”ëœ ì¢Œí‘œ(0~1)ë¥¼ í”½ì…€ ì¢Œí‘œë¡œ ë³€í™˜
                x1, y1 = int(pt1.x * width), int(pt1.y * height)
                x2, y2 = int(pt2.x * width), int(pt2.y * height)
                
                # ì§§ì€ ì„ ë¶„(Segment) í˜•íƒœë¡œ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
                # [[[x1, y1]], [[x2, y2]]] í˜•íƒœëŠ” cv2.drawContoursì™€ í˜¸í™˜ë©ë‹ˆë‹¤.
                segment = np.array([[[x1, y1]], [[x2, y2]]], dtype=np.int32)
                final_contours.append(segment)

    print(f"âœ… ì–¼êµ´ íŠ¹ì§• ì¶”ì¶œ ì™„ë£Œ: ì´ {len(final_contours)}ê°œì˜ ì„ ë¶„ ìƒì„±")
    
    # (ì˜µì…˜) ë””ë²„ê¹…ìš© ì´ë¯¸ì§€ ì €ì¥
    debug_img = np.zeros((height, width), dtype=np.uint8)
    cv2.drawContours(debug_img, final_contours, -1, 255, 1)
    cv2.imwrite("debug_face_result.png", debug_img)
    print("   ã„´ í™•ì¸ìš© ì´ë¯¸ì§€ ì €ì¥: debug_face_result.png")
    
    return final_contours

# =========================================================
# [Block 3] G-code ë° SVG ìƒì„±
# =========================================================
def save_gcode_and_svg(contours, width, height):
    """
    ì¶”ì¶œëœ Contoursë¥¼ G-code(.nc)ì™€ SVG íŒŒì¼ë¡œ ì €ì¥
    """
    print(f"ğŸ”„ [3ë‹¨ê³„] íŒŒì¼ ìƒì„± ì‹œì‘ ({NC_FILENAME}, {SVG_FILENAME})...")
    
    dwg = svgwrite.Drawing(SVG_FILENAME, profile='tiny', size=(width, height))
    
    count = 0
    with open(NC_FILENAME, 'w') as f:
        # G-code í—¤ë”
        f.write("G21\nG90\nG0 Z10.0\n") # mmë‹¨ìœ„, ì ˆëŒ€ì¢Œí‘œ, ì•ˆì „ë†’ì´ ì´ë™
        
        for contour in contours:
            # ì ë“¤ì˜ ë°°ì—´ë¡œ ë³€í™˜ (Flatten)
            points = contour.reshape(-1, 2)
            
            if len(points) < 2: continue
            
            # --- SVG ì €ì¥ ---
            # numpy int32 -> python native float ë³€í™˜ í•„ìš”
            svg_data = [(float(p[0]), float(p[1])) for p in points]
            dwg.add(dwg.polyline(svg_data, stroke='black', fill='none', stroke_width=1))
            
            # --- G-code ì €ì¥ ---
            # 1. ì‹œì‘ì ìœ¼ë¡œ ì´ë™ (íœ ë“¤ê³ )
            start_x = float(points[0][0]) * SCALE
            start_y = float(points[0][1]) * SCALE
            f.write(f"G0 X{start_x:.3f} Y{-start_y:.3f}\n")
            
            # 2. ê·¸ë¦¬ê¸° ì‹œì‘ (íœ ë‚´ë¦¬ê³ )
            f.write(f"G1 Z{Z_DRAW} F{FEED_RATE}\n")
            
            # 3. ê²½ë¡œ ë”°ë¼ ì´ë™
            for i in range(1, len(points)):
                px = float(points[i][0]) * SCALE
                py = float(points[i][1]) * SCALE
                f.write(f"G1 X{px:.3f} Y{-py:.3f}\n")
            
            # 4. íš ëë‚¨ (íœ ë“¤ê¸°)
            f.write(f"G0 Z{Z_SAFE}\n")
            count += 1
            
        # G-code í‘¸í„°
        f.write("G0 X0 Y0\nM2\n")
        
    dwg.save()
    print(f"âœ… ëª¨ë“  ì‘ì—… ì™„ë£Œ! ì´ {count}ê°œì˜ íšì´ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

# =========================================================
# [ë©”ì¸ ì‹¤í–‰ë¶€]
# =========================================================
if __name__ == "__main__":
    # ì‚¬ìš©ì ì…ë ¥
    input_filename = input("íŒŒì¼ ëª…ì„ ì…ë ¥í•˜ì„¸ìš” (í™•ì¥ì í¬í•¨, ì˜ˆ: face.jpg) : ")
    
    # 1. ë°°ê²½ ì œê±° ë° ë¡œë“œ
    img_processed = load_and_remove_bg(input_filename)
    
    if img_processed is not None:
        # 2. AIë¡œ ì–¼êµ´ ì„  ì¶”ì¶œ
        contours = extract_face_contours_mediapipe(img_processed)
        
        if contours:
            # 3. íŒŒì¼ ì €ì¥
            h, w = img_processed.shape[:2]
            save_gcode_and_svg(contours, w, h)
```
